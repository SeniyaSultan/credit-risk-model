```python
import pandas as pd
import zipfile

zip_file = "data (2).zip"

with zipfile.ZipFile(zip_file) as z:
    with z.open("data.csv") as f:
        df = pd.read_csv(f)

df.head()
```

```python
num_cols = ["Amount", "Value"]

for col in num_cols:
    plt.figure()
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
cat_cols = [
    "CurrencyCode", "CountryCode", "ChannelId",
    "ProductCategory", "ProviderId"
]

for col in cat_cols:
    print(df[col].value_counts().head(10))
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
df.isnull().sum()
```

```python
for col in num_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

## Key EDA Insights

1. Transaction amounts are highly skewed, with a small number of customers contributing very large transaction values. This suggests the need for scaling and possibly log transformations during feature engineering.

2. Customer behavior varies significantly across channels and product categories, indicating that categorical features such as ChannelId and ProductCategory may be strong predictors of credit risk.

3. Many customers have low transaction frequency and low monetary value, supporting the use of RFM analysis to identify disengaged and potentially high-risk customers.

4. The dataset contains minimal missing values, making imputation feasible without significant data loss.

5. Clear temporal information exists in the transaction timestamps, enabling time-based feature extraction and recency analysis.
